```python
import pandas as pd
```

```python
import numpy as np
```

```python
# read csv
tweets_df = pd.read_csv("processed_tweet_data.csv")
tweets_df
```

```python
# select necessary columns
cleanTweet = tweets_df[["clean_text", "polarity"]]
cleanTweet
```

```python
def text_category(p):
    """
    converts polarity into sentiment category
    """
    if p > 0:
        return "positive"
    elif p < 0:
        return "negative"
    else:
        return "neutral"
```

```python
cleanTweet["score"] = cleanTweet["polarity"].apply(text_category)
cleanTweet
```

```python
sentiment = cleanTweet['score'].value_counts()
sentiment
```
